```python
import numpy as np
import pandas as pd
from PIL import Image
import glob
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
def get_info(filenames,classes):
    lst=[]
    for file in filenames:
        img=Image.open(file)
        r=np.asarray(img)[:,:,0]
        g=np.asarray(img)[:,:,1]
        b=np.asarray(img)[:,:,2]

        rmean=r.mean()
        gmean=g.mean()
        bmean=b.mean()

        rstd=r.std()
        gstd=g.std()
        bstd=b.std()


        rqua=np.quantile(r, 0.25)
        gqua=np.quantile(g, 0.25)
        bqua=np.quantile(b, 0.25)

        rmed=np.median(r)
        gmed=np.median(g)
        bmed=np.median(b)
        
        rqua75=np.quantile(r, 0.75)
        gqua75=np.quantile(g, 0.75)
        bqua75=np.quantile(b, 0.75)
        
        r_r_mean=r[:int(len(r)*0.5)].mean()-r[-int(len(r)*0.5):].mean()
        g_g_mean=g[:int(len(g)*0.5)].mean()-g[-int(len(g)*0.5):].mean()
        b_b_mean=r[:int(len(b)*0.5)].mean()-b[-int(len(b)*0.5):].mean()
        #можно все заменить на pd.DataFrame(np.asarray(img)[:,:,0].reshape(-1)).describe()
        lst.append([rmean,gmean,bmean,rstd,gstd,bstd,rqua,gqua,bqua,rmed,gmed,bmed,rqua75,gqua75,bqua75,r_r_mean,g_g_mean,b_b_mean,classes])
    return lst
```

```python
ind=get_info(glob.glob('train\\indoor\\*'),1)
outd=get_info(glob.glob('train\\outdoor\\*'),0)
```

```python
df = pd.DataFrame(np.vstack((ind,outd)), columns = ['rmean','gmean','bmean','rstd','gstd','bstd','rqua','gqua','bqua','rmed','gmed','bmed','rqua75','gqua75','bqua75','r_r_mean','g_g_mean','b_b_mean','class'])
```

```python
df_new=df.sample(frac=1)
df_new
```

```python
X=df_new[np.array(df.columns[:-1])]
y=df_new[np.array(df.columns[-1])]
```

```python
X_train=X[:int(X.shape[0]*0.6)]
X_test=X[:int(X.shape[0]*0.3)]
y_train=y[:int(y.shape[0]*0.6)]
y_test=y[:int(y.shape[0]*0.3)]
```

```python
clf = LogisticRegression(random_state=0).fit(X_train, y_train)
y_pred=clf.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```
