```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('golf_binary.csv', index_col=0)
```

```python
df['Outlook']=df.index
df=df.set_index(pd.Index(list(range(df.shape[0]))))
```

```python
df
```

```python
uniq=[df[i].unique().tolist() for i in df.columns]
len_of_uniq=list(map(len,uniq))
```

```python
uniq
```

```python
new_df=pd.DataFrame()
for i in range(len(uniq)):
    if len_of_uniq[i]==2:
        new_df[df.columns[i]]=df[df.columns[i]].replace({uniq[i][1]:1,uniq[i][0]:0})
    if len_of_uniq[i]==3:
        for j in range(3):
            lst=uniq[i][:j] + uniq[i][j+1:]
            new_df[uniq[i][j]]=df[df.columns[i]].replace({uniq[i][j]:1,lst[0]:0,lst[1]:0})
```

```python
new_df
```

```python
probability=[]
len_new_df=new_df.shape[0]
for i in range(len_new_df):

    temperature=new_df.iloc[i][['Hot','Mild','Cool']].values
    outlook=new_df.iloc[i][['Rainy', 'Overcast', 'Sunny']].values
    humidity=new_df.iloc[i]['Humidity']
    y=new_df.iloc[i]['Play Golf']
    wind=new_df.iloc[i]['Wind']

    y1=new_df[new_df['Play Golf']==y].shape[0]
    x1=new_df[(new_df['Play Golf']==y) & (new_df['Humidity']==humidity)].shape[0]/y1
    x2=new_df[(new_df['Play Golf']==y) & (new_df['Wind']==wind)].shape[0]/y1
    x3=new_df[(new_df['Play Golf']==y) & (new_df['Hot']==temperature[0]) & (new_df['Mild']==temperature[1]) & (new_df['Cool']==temperature[2])].shape[0]/y1
    x4=new_df[(new_df['Play Golf']==y) & (new_df['Rainy']==outlook[0]) & (new_df['Overcast']==outlook[1]) & (new_df['Sunny']==outlook[2])].shape[0]/y1

    y1_1=new_df[new_df['Play Golf']==1-y].shape[0]
    x1_1=new_df[(new_df['Play Golf']==1-y) & (new_df['Humidity']==humidity)].shape[0]/y1_1
    x2_1=new_df[(new_df['Play Golf']==1-y) & (new_df['Wind']==wind)].shape[0]/y1_1
    x3_1=new_df[(new_df['Play Golf']==1-y) & (new_df['Hot']==temperature[0]) & (new_df['Mild']==temperature[1]) & (new_df['Cool']==temperature[2])].shape[0]/y1_1
    x4_1=new_df[(new_df['Play Golf']==1-y) & (new_df['Rainy']==outlook[0]) & (new_df['Overcast']==outlook[1]) & (new_df['Sunny']==outlook[2])].shape[0]/y1_1
    
    probability.append((y1/len_new_df*x1*x2*x3*x4)/(y1/len_new_df*x1*x2*x3*x4+y1_1/len_new_df*x1_1*x2_1*x3_1*x4_1))
```

```python
new_df['Probability']=probability
new_df
```

```python
df['Probability']=probability
df # в данном лсучае оценена вероятность наборов 
```

```python
def find_prob(temperature,outlook,humidity,wind):
    len_new_df=new_df.shape[0]
    y=1

    y1=new_df[new_df['Play Golf']==y].shape[0]
    x1=new_df[(new_df['Play Golf']==y) & (new_df['Humidity']==humidity)].shape[0]/y1
    x2=new_df[(new_df['Play Golf']==y) & (new_df['Wind']==wind)].shape[0]/y1
    x3=new_df[(new_df['Play Golf']==y) & (new_df['Hot']==temperature[0]) & (new_df['Mild']==temperature[1]) & (new_df['Cool']==temperature[2])].shape[0]/y1
    x4=new_df[(new_df['Play Golf']==y) & (new_df['Rainy']==outlook[0]) & (new_df['Overcast']==outlook[1]) & (new_df['Sunny']==outlook[2])].shape[0]/y1

    y1_1=new_df[new_df['Play Golf']==1-y].shape[0]
    x1_1=new_df[(new_df['Play Golf']==1-y) & (new_df['Humidity']==humidity)].shape[0]/y1_1
    x2_1=new_df[(new_df['Play Golf']==1-y) & (new_df['Wind']==wind)].shape[0]/y1_1
    x3_1=new_df[(new_df['Play Golf']==1-y) & (new_df['Hot']==temperature[0]) & (new_df['Mild']==temperature[1]) & (new_df['Cool']==temperature[2])].shape[0]/y1_1
    x4_1=new_df[(new_df['Play Golf']==1-y) & (new_df['Rainy']==outlook[0]) & (new_df['Overcast']==outlook[1]) & (new_df['Sunny']==outlook[2])].shape[0]/y1_1
    
    pr=((y1/len_new_df)*x1*x2*x3*x4)/((y1/len_new_df)*x1*x2*x3*x4+(y1_1/len_new_df)*x1_1*x2_1*x3_1*x4_1)
    return [pr,1-pr]
```
